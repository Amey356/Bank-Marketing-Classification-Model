```python
cat = []
num = []


for i in df_resampled.columns:
    if df_resampled[i].dtype == 'O':
        cat.append(i)
    else:
        num.append(i)
```

```python
# Cat

for i in cat:
    vc = df_resampled[i].value_counts()
    plt.bar(vc.index, height = vc , color = 'b')
    plt.xlabel(i)
    plt.xticks(rotation = 90)
    plt.ylabel('Count')
    plt.show()
```

```python
df_resampled.drop(['default','loan','contact','poutcome'], axis = 1, inplace = True)
```

```python
# Numerical

for i in num:
    fig, ax = plt.subplots(1,2)
    ax[0].hist(df_resampled[i],color = 'c')
    ax[1].boxplot(df_resampled[i])
    plt.suptitle(i)
    plt.show()
```

```python
# iqr winsorization --
# replace lower outliers with lower fence
# replace upper outliers with upper fence


for i in num:
    q1 = df_resampled[i].quantile(0.25)
    q3 = df_resampled[i].quantile(0.75)
    iqr = q3 - q1
    lf = q1 - 1.5*iqr
    uf = q3 + 1.5*iqr
    df_resampled.loc[df_resampled[i]<lf, i] = lf
    df_resampled.loc[df_resampled[i]>uf, i] = uf
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, confusion_matrix

# Training the model
model = LogisticRegression(random_state=42, class_weight="balanced", solver="liblinear", max_iter=500)
model.fit(x_train, y_train)

# Predicting on training and testing data
y_train_predicted = model.predict(x_train)
y_test_predicted = model.predict(x_test)

# Training Performance
print("Training Performance: ")
print("Training accuracy: ", accuracy_score(y_train, y_train_predicted))
print("Training precision: ", precision_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Training recall: ", recall_score(y_train, y_train_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_train, y_train_predicted))
print("Training classification report: ", classification_report(y_train, y_train_predicted))

# Testing Performance
print("Testing Performance: ")
print("Testing accuracy: ", accuracy_score(y_test, y_test_predicted))
print("Testing precision: ", precision_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Testing recall: ", recall_score(y_test, y_test_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_test, y_test_predicted))
print("Testing classification report: ", classification_report(y_test, y_test_predicted))
```

# KNN Regression

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Initialize the KNN model
knn_model = KNeighborsClassifier(metric='manhattan', n_neighbors=9, weights='uniform')
knn_model.fit(x_train, y_train) 


# Predict on training data
ytrain_knn_predicted = knn_model.predict(x_train)

# Training Performance
print("Training Performance: ")
print("Training accuracy: ", accuracy_score(y_train, ytrain_knn_predicted))
print("Training precision: ", precision_score(y_train, ytrain_knn_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Training recall: ", recall_score(y_train, ytrain_knn_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_train, ytrain_knn_predicted))
print("Training classification report: ", classification_report(y_train, ytrain_knn_predicted))

# Predict on testing data
y_test_knn_predicted = knn_model.predict(x_test)

# Testing Performance
print("Testing Performance: ")
print("Testing accuracy: ", accuracy_score(y_test, y_test_knn_predicted))
print("Testing precision: ", precision_score(y_test, y_test_knn_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print("Testing recall: ", recall_score(y_test, y_test_knn_predicted, average='binary'))  # 'binary' or 'macro' for multi-class
print(confusion_matrix(y_test, y_test_knn_predicted))
print("Testing classification report: ", classification_report(y_test, y_test_knn_predicted))
```

## Grid search cv - KNN

```python
from sklearn.model_selection import GridSearchCV

clf = KNeighborsClassifier()

param_grid = {
    'n_neighbors': [3, 5, 7, 9, 11],
    'weights': ['uniform', 'distance'],
    'metric': ['euclidean', 'manhattan', 'minkowski']
}

grid_search = GridSearchCV(estimator=clf, param_grid=param_grid, cv=5, scoring='accuracy', n_jobs=-1, verbose=1)
grid_search.fit(x_train, y_train)
print("Best Parameters:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)
best_model = grid_search.best_estimator_
test_score = best_model.score(x_test, y_test)
print("Test Accuracy:", test_score)
```

# Voting Classifier

```python
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, confusion_matrix, classification_report

# Initialize the base models
model1 = LogisticRegression()
model2 = DecisionTreeClassifier()
model3 = SVC()

# Create the ensemble classifier using VotingClassifier
ensemble_classifier = VotingClassifier(
    estimators=[('lr', model1), ('dt', model2), ('svc', model3)], 
    voting='hard'  # 'hard' for majority voting
)

# Train the ensemble classifier on the training data
ensemble_classifier.fit(x_train, y_train)

# Predict on training and testing data
y_train_predicted = ensemble_classifier.predict(x_train)
y_test_predicted = ensemble_classifier.predict(x_test)

# Evaluate the model on training data
print("Training Model Performance:")
print("Training Accuracy:", accuracy_score(y_train, y_train_predicted))
print("Training Precision:", precision_score(y_train, y_train_predicted, average='binary'))  # Change to 'macro' or 'weighted' for multi-class
print("Training Recall:", recall_score(y_train, y_train_predicted, average='binary'))  # Change to 'macro' or 'weighted' for multi-class
print("Training Confusion Matrix:\n", confusion_matrix(y_train, y_train_predicted))
print("Training Classification Report:\n", classification_report(y_train, y_train_predicted))

# Evaluate the model on testing data
print("\nTesting Model Performance:")
print("Testing Accuracy:", accuracy_score(y_test, y_test_predicted))
print("Testing Precision:", precision_score(y_test, y_test_predicted, average='binary'))  # Change to 'macro' or 'weighted' for multi-class
print("Testing Recall:", recall_score(y_test, y_test_predicted, average='binary'))  # Change to 'macro' or 'weighted' for multi-class
print("Testing Confusion Matrix:\n", confusion_matrix(y_test, y_test_predicted))
print("Testing Classification Report:\n", classification_report(y_test, y_test_predicted))
```
